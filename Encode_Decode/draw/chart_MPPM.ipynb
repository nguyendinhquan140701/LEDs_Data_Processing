```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager, rcParams
print(rcParams['font.family'])
```

```python
path_file2 = "E:\Project\OOC\LEDs_Data_Processing\Encode_Decode\draw\data\mppm-copy.xlsx" ## for the 2 LED
path_file1 = "E:\Project\OOC\LEDs_Data_Processing\Encode_Decode\draw\data\mppm.xlsx"  ## for the 1 LED

df1 = pd.read_excel(path_file1)
df2 = pd.read_excel(path_file2)

# print(df)
```

```python
distance = df1.iloc[: 10, 0]
# print(distance)

c1_14khz_accuracy= df1.iloc[:10, 1].to_list()
c1_18khz_accuracy= df1.iloc[10:20, 1].to_list()
c1_20khz_accuracy= df1.iloc[20:30, 1].to_list()

# print(distance)
c2_14khz_accuracy= df2.iloc[:10, 2].to_list()
c2_18khz_accuracy= df2.iloc[10:20, 2].to_list()
c2_20khz_accuracy= df2.iloc[20:30, 2].to_list()


c3_14khz_accuracy= df2.iloc[:10, 3].to_list()
c3_18khz_accuracy= df2.iloc[10:20, 3].to_list()
c3_20khz_accuracy= df2.iloc[20:30, 3].to_list()


_14khz_ber = df1.iloc[:10, 5].to_list()
_18khz_ber = df1.iloc[10:20, 5].to_list()
_20khz_ber = df1.iloc[20:30, 5].to_list()

_2LED_14khz_ber = df2.iloc[:10, 5].to_list()
_2LED_18khz_ber = df2.iloc[10:20, 5].to_list()
_2LED_20khz_ber = df2.iloc[20:30, 5].to_list()
# print(c3_20khz_accuracy)

```

1 led 14khz

```python
from matplotlib import font_manager


rcParams['font.family'] = 'Times New Roman'
# Define font properties for Times New Roman
legend_font = font_manager.FontProperties( style='normal', weight='bold', size=5)
label_font = {'style': 'normal', 'weight': 'bold', 'size': 10}

fig, ax1 = plt.subplots(figsize=(2, 2))

ax1.set_xlabel('Distance (m)', fontsize=10, fontdict=label_font)
ax1.tick_params(axis='x', labelcolor='black', labelsize=10, pad = 3)
ax1.set_xticks([1.5, 1.7, 1.9, 2.1, 2.3])

ax1.set_ylabel('Decodable frame rate (%)', color='black', fontsize=10, fontdict=label_font)
# ax1.set_xlim(1.6, 2)
ax1.set_ylim(-2, 100)
ax1.tick_params(axis='y', labelcolor='black', labelsize=10, pad = 3)
ax1.set_yticks([0, 20, 40, 60, 80, 100])
line1 = ax1.plot(distance, c1_14khz_accuracy, color='Blue', label='DFR of code 1 (%)', marker='o', markersize=2, linewidth=1)

# ax1.scatter(distance, c1_code1_accuracy, color='blue', edgecolors='blue', s=3)
for i, txt in enumerate(c1_14khz_accuracy):
    if i in [0, 2, 4, 6]:
        ax1.annotate(f'{txt:.2f}', (distance[i], c1_14khz_accuracy[i]), textcoords="offset points", 
                     xytext=(6,4), ha='center', color='black', fontsize=6, fontstyle='normal')

ax2 = ax1.twinx()
ax2.set_ylim(-0.16, 6)
ax2.set_ylabel('Bit Error Rate (%)', color='black', fontsize=10, fontdict=label_font)
ax2.tick_params(axis='y', labelcolor='black', labelsize=10, pad = 3)
line3 = ax2.plot(distance, _14khz_ber, color='red', label='BER (%)', linestyle='--', marker='o', markersize=2, linewidth=1)


peak_index = np.argmax(_14khz_ber)  # Index of the peak point
peak_value = _14khz_ber[peak_index]  # Value of the peak point
peak_distance = distance[peak_index]  # Distance of the peak point
ax2.annotate(f'Max = {peak_value:.2f}', 
             (peak_distance, peak_value), 
             textcoords="offset points", 
             xytext=(0, 6), 
             ha='center', 
             color='red', 
             fontsize=5, 
             fontstyle='normal', 
             fontweight='bold',
             family='Times New Roman',
             arrowprops=dict(facecolor='red', edgecolor='none', shrink=0.1, width=2, headwidth=3, headlength=4))

lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]

# Move the legend above the plot, outside the graph
ax1.legend(lines, labels, loc='upper center', bbox_to_anchor=(0.5, 1.17), ncol=2, prop=legend_font)

plt.savefig('E:\Project\OOC\LEDs_Data_Processing\Encode_Decode\draw\img\_1led_14khz.jpeg', dpi = 600,bbox_inches='tight')
```
