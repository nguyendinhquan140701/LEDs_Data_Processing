```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img_path = "../led_image/paper/paper_4_LED.png"
img = cv2.imread(img_path)
height, width, _ = img.shape
print(f'height:{height}, width:{width}')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY +
    cv2.THRESH_OTSU)
# show thresholded image
# plt.imshow(thresh, cmap='gray')
# plt.show()
```

```python
contours, hierarchy = cv2.findContours(
    thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE
)
boundingBox = []
for contour in contours:
    # area = cv2.contourArea(contour)
    x, y, w, h = cv2.boundingRect(contour)
    boundingBox.append((x, y, w, h))
    print("boundingBox = ", boundingBox)
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 1)

x = []
y = []
w = []
h = []
for i in range (0, len(boundingBox)):
    x.append(boundingBox[i][0])
    y.append(boundingBox[i][1])
    w.append(boundingBox[i][2])
    h.append(boundingBox[i][3])
print( "w = ", w)

h_max = max(boundingBox[:][3])
if h_max != 0:
    print("h_max = ", h_max)
    print("y:", y)
# img = cv2.resize(img, (500, 500))
plt.imshow(img, cmap='gray')
cv2.imwrite('input_paper3.png', img)
```
